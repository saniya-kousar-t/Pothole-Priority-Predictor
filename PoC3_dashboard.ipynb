```python
import json
import matplotlib.pyplot as plt
import numpy as np

# Hardcoded JSON data
data = json.loads('''
{
    "model_metrics": {
        "mAP50": 0.883,
        "mAP50_95": 0.658,
        "Precision": 0.861,
        "Recall": 0.810
    },
    "dataset_distribution": {
        "Train": 6809,
        "Validation": 1944,
        "Test": 974
    },
    "detection_summary": {
        "Total Images": 9727,
        "Total Potholes Detected": 4829,
        "Avg Potholes Per Image": 1.78
    }
}
''')

# Dashboard with 3 charts
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Pothole Priority Predictor - Dashboard', fontsize=16, fontweight='bold')

# Chart 1 - Model Metrics
metrics = data["model_metrics"]
axes[0].bar(metrics.keys(), metrics.values(), color=['blue','orange','green','red'])
axes[0].set_title('Model Performance Metrics')
axes[0].set_ylim(0, 1)
axes[0].set_ylabel('Score')
for i, v in enumerate(metrics.values()):
    axes[0].text(i, v + 0.01, f'{v:.3f}', ha='center')

# Chart 2 - Dataset Distribution
dist = data["dataset_distribution"]
axes[1].pie(dist.values(), labels=dist.keys(), autopct='%1.1f%%', colors=['blue','orange','green'])
axes[1].set_title('Dataset Distribution')

# Chart 3 - Detection Summary
summary = data["detection_summary"]
axes[2].bar(summary.keys(), summary.values(), color='purple')
axes[2].set_title('Detection Summary')
for i, v in enumerate(summary.values()):
    axes[2].text(i, v + 10, str(v), ha='center')

plt.tight_layout()
plt.savefig(r"C:\Users\Dell\Desktop\dashboard_poc3.png")
plt.show()
print("PoC 3 Dashboard rendered successfully!")
```
