```python
import os
import glob
import matplotlib.pyplot as plt

# Correct path
base = r"C:\Users\Dell\Desktop\final_dataset"

train_imgs = glob.glob(base + r"\train\images\*")
valid_imgs = glob.glob(base + r"\valid\images\*")
test_imgs  = glob.glob(base + r"\test\images\*")

print(f"Train images:      {len(train_imgs)}")
print(f"Validation images: {len(valid_imgs)}")
print(f"Test images:       {len(test_imgs)}")
print(f"Total images:      {len(train_imgs)+len(valid_imgs)+len(test_imgs)}")

# Class distribution from labels
train_labels = glob.glob(base + r"\train\labels\*.txt")
total_potholes = 0
for label in train_labels:
    with open(label) as f:
        lines = f.readlines()
        total_potholes += len(lines)

print(f"\nClass: Pothole")
print(f"Total pothole annotations in train: {total_potholes}")
if len(train_imgs) > 0:
    print(f"Avg potholes per image: {total_potholes/len(train_imgs):.2f}")

# Plot
labels = ['Train', 'Validation', 'Test']
counts = [len(train_imgs), len(valid_imgs), len(test_imgs)]
plt.bar(labels, counts, color=['blue', 'orange', 'green'])
plt.title('Dataset Distribution')
plt.ylabel('Number of Images')
plt.show()
```

```python

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax = plt.subplots(figsize=(18, 10))
ax.set_xlim(0, 18)
ax.set_ylim(0, 10)
ax.axis('off')
ax.set_facecolor('#1a1a2e')
fig.patch.set_facecolor('#1a1a2e')
ax.set_title('Pothole Priority Predictor — Complete System Architecture',
             fontsize=15, fontweight='bold', color='white', pad=20)

# Define layers
layers = [
    # (x, y, w, h, label, color)
    # Input Layer
    (0.3, 7.5, 2.2, 1.2, '📷 Camera\nInput', '#4A90D9'),
    (0.3, 5.5, 2.2, 1.2, '🖼️ Image\nUpload', '#4A90D9'),

    # Processing Layer
    (3.5, 7.5, 2.2, 1.2, '⚙️ Preprocessing\n(Resize 640x640)', '#7B68EE'),
    (3.5, 5.5, 2.2, 1.2, '📦 Roboflow\nDataset', '#7B68EE'),

    # ML Layer
    (7.0, 6.5, 2.5, 1.5, '🧠 YOLOv11\nDetection Model', '#E67E22'),

    # Scoring Layer
    (10.5, 6.5, 2.2, 1.5, '⚠️ Severity\nScoring\n(Size+Count)', '#E74C3C'),

    # Backend Layer
    (13.5, 7.5, 2.2, 1.2, '🔧 Flask\nAPI', '#27AE60'),
    (13.5, 5.5, 2.2, 1.2, '🗄️ Database\n(Results)', '#27AE60'),

    # Output Layer
    (16.3, 7.5, 1.5, 1.2, '📊 Streamlit\nDashboard', '#E74C3C'),
    (16.3, 5.5, 1.5, 1.2, '📋 Priority\nReport', '#E74C3C'),
]

for x, y, w, h, label, color in layers:
    rect = mpatches.FancyBboxPatch((x, y), w, h,
        boxstyle="round,pad=0.15", linewidth=2,
        edgecolor='white', facecolor=color, alpha=0.9)
    ax.add_patch(rect)
    ax.text(x + w/2, y + h/2, label, ha='center', va='center',
            fontsize=9, fontweight='bold', color='white')

# Layer Headers
headers = [
    (1.4, 9.2, 'INPUT\nLAYER', '#4A90D9'),
    (4.6, 9.2, 'PROCESSING\nLAYER', '#7B68EE'),
    (8.25, 9.2, 'ML\nLAYER', '#E67E22'),
    (11.6, 9.2, 'SCORING\nLAYER', '#E74C3C'),
    (14.6, 9.2, 'BACKEND\nLAYER', '#27AE60'),
    (17.0, 9.2, 'OUTPUT\nLAYER', '#E74C3C'),
]

for x, y, label, color in headers:
    ax.text(x, y, label, ha='center', va='center', fontsize=8,
            fontweight='bold', color=color)

# Arrows
arrows = [
    (2.5, 8.1, 3.5, 8.1),   # Camera → Preprocess
    (2.5, 6.1, 3.5, 6.1),   # Upload → Dataset
    (5.7, 8.1, 7.0, 7.25),  # Preprocess → YOLO
    (5.7, 6.1, 7.0, 7.0),   # Dataset → YOLO
    (9.5, 7.25, 10.5, 7.25),# YOLO → Severity
    (12.7, 7.25, 13.5, 8.1),# Severity → Flask
    (12.7, 7.0, 13.5, 6.1), # Severity → DB
    (15.7, 8.1, 16.3, 8.1), # Flask → Streamlit
    (15.7, 6.1, 16.3, 6.1), # DB → Report
]

for x1, y1, x2, y2 in arrows:
    ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
                arrowprops=dict(arrowstyle='->', color='white', lw=1.5))

# Metrics box at bottom
ax.text(9, 1.5,
        'Model: YOLOv11n  |  Dataset: 9727 images  |  mAP50: 0.883  |  Precision: 0.861  |  Recall: 0.810',
        ha='center', va='center', fontsize=10, color='white',
        bbox=dict(boxstyle='round', facecolor='#2ECC71', alpha=0.3))

# Severity legend
ax.text(9, 0.7, 'Severity Levels:  🔴 High (>5 potholes)   🟡 Medium (2-5)   🟢 Low (1)',
        ha='center', va='center', fontsize=10, color='white',
        bbox=dict(boxstyle='round', facecolor='#E74C3C', alpha=0.2))

plt.tight_layout()
plt.savefig(r"C:\Users\Dell\Desktop\architecture_diagram.png", dpi=150, bbox_inches='tight')
plt.show()
print("Complete architecture diagram saved!")
```
